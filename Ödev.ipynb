```python
def deneme(a, b):
    return a - b
```

```python
df['Netfiyat'] = df.apply(lambda row: deneme(row['total_bill'], row['tip']), axis=1)
```

```python
df.head(5)
```

```python
sns.swarmplot(x="smoker", y="tip", hue="sex",data=df,) # Farklı bir yaklaşım olarak kategorik eksendeki her dağılma noktasını, noktaların çakışmasını önleyen bir algoritmayla 
                                                       #konumlandıran swarmplot() fonksiyonunu kullanmak olacaktır 
```

```python
sns.countplot(x="smoker", hue="sex", data=df,palette="magma");#İkinci bir değişken için bir istatistik hesaplamaktan ziyade, her kategorideki gözlemlerin hiçbirinin gösterilmemesidir.
                                                              #Bunun için bu fonksiyonu kullanıyoruz
```

```python
sns.relplot(x="total_bill", y="tip",hue="sex",style="time",palette="magma",data=df);
```

```python
sns.relplot(x="day", y="tip", hue="sex",                      #Bir FacetGrid üzerine ilişkisel çizimler çizme için kullanılır.
            dashes=True, markers=False, kind="line", data=df);#Bu işlev, alt kümelerin semantik eşlemeleriyle iki değişken arasındaki ilişkiyi gösteren birkaç farklı eksen seviyesi 
                                                              #işlevine erişim sağlar
```

```python
tita.groupby(['survived']).size()
```

```python
sns.factorplot("class", "survived", "sex", data=tita, kind="bar", palette="magma", legend=False)
#Bir FacetGrid üzerinde kategorik bir çizim çizer. 'kind' parametresini kullanarak boxplot, violinplot, barplot 
#ve stripplot gibi türleri seçebiliriz. FacetGrid varsayılan olarak pointplot kullanır.
```

```python
sns.boxplot(x = "sex", y = "tip",data = df) #Bu tür bir grafik, dağıtımın çeyrek değerlerini aşırı değerler ile birlikte göstermektedir.
```

```python
sns.barplot(x = "sex", y = "survived", hue = "class", data = tita) 
# Bu fonksiyonda her sınıftaki erkek ve dişi ortalama hayatta kalma sayısını görebiliriz
```

```python
sns.pointplot(x = "sex", y = "Netfiyat", hue = "smoker", data = df)
#Çubuk çizimleriyle aynıdır ancak farklı bir tarzda. Tam çubuktan ziyade, tahminin değeri,
#diğer eksen üzerindeki belirli bir yükseklikteki nokta ile temsil edilir.
```

```python
df['tip_orn'] = 100 * df['tip'] / df['total_bill']

grid = sns.FacetGrid(df, row="sex", col="time", margin_titles=True)
grid.map(plt.hist, "tip_orn", bins=np.linspace(0, 40, 15));#Bir histogram kullanarak bu alt kümelerin her birinde ipuçlarının dağılımını gösterd0ik
#fdeğişkenleri bölerek satır ve sütun tarafından tanımlanan bir panel matrisini oluşturur. 
#Paneller nedeniyle, tek bir plot birden fazla panele benziyor. İki ayrı değişkende tüm kombinasyonları analiz etmek çok faydalıdır.
```

```python
sns.jointplot("total_bill", "tip", data=df, kind='hex')
#Marjinal tek değişkenli plot ile iki değişkenli bir çizim çizmek için kullanılır.
```
